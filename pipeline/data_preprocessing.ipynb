```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\Summerbreak\Scrapping_op.gg\match_data.csv")
print(df.head)
```

```python
df["team1_champs"]= df["team1"].apply(lambda x:x.split(','))
df["team2_champs"]= df["team2"].apply(lambda x:x.split(','))

print(df.head)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
all_champs_list = df['team1_champs'].tolist() + df['team2_champs'].tolist()
mlb = MultiLabelBinarizer()
mlb.fit(all_champs_list)
```

```python

team1_encoded = pd.DataFrame(
    mlb.transform(df['team1_champs']),
    columns=[f't1_{champ}' for champ in mlb.classes_]
)
team2_encoded = pd.DataFrame(
    mlb.transform(df['team2_champs']),
    columns=[f't2_{champ}' for champ in mlb.classes_]
)
```

```python
df_encoded = pd.concat([df, team1_encoded, team2_encoded], axis=1)
df_encoded.drop(['team1', 'team2', 'team1_champs', 'team2_champs'], axis=1, inplace=True)
df_encoded.to_csv('meow.csv', index=False)
```
